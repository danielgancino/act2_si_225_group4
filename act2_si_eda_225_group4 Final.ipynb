###3. Delete columns that are not relevant or useful for the analysis [Daniel]

```python
df.drop(['username', 'room_view', 'review_title', 'original_lang', 'owner_resp_text', 'en_full_review', 'review_text_liked', 'review_text_disliked'], axis=1, inplace=True)
df
```

**Justification:**

After exploring the dataset, several columns were found irrelevant for analyzing hotel reviews. **Username** was removed as it only identifies the reviewer, and **room_view** because it does not reflect overall satisfaction. **Review_title** was excluded for being redundant with the full review, while **original_lang** was unnecessary since an English-translated version already exists. Finally, **owner_resp_text** was discarded as it represents the hotel’s reply rather than guest feedback.
This reduction leaves the dataset cleaner and focused on variables that better explain guest satisfaction and review trends.

###5. Find the column related to the review date and splitted into two columns [Antonio]

```python
df['review_post_date'] = pd.to_datetime(df['review_post_date'], format="%m-%d-%Y %H:%M:%S")

df['review_year'] = df['review_post_date'].dt.year
df['review_month'] = df['review_post_date'].dt.month

df = df[[
    "user_country", "stay_duration", "stay_type",
    "review_post_date", "review_year", "review_month",
    "rating", "full_review"
]]

df
```

###10. Graphs [Daniel, Dylan, Antonio]

```python
rating = df.groupby('rating').size()
rating

df['rating'].value_counts().reindex(range(1,11), fill_value=0).plot(kind='bar', title='Rating', xlabel='Rating', ylabel='Quantity', color='orange')
plt.xticks(rotation=0)
plt.show()
```

------------------
Reviews per month [Dylan]

```python
months_labels = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dic']
df['review_month'].value_counts().reindex(range(1,13), fill_value=0).plot(kind='bar', title='Reviews per Month', xlabel='Month', ylabel='Amount')
plt.xticks(ticks=range(12), labels=months_labels, rotation=0)
plt.show()
```

----------------
Reviews per year [Daniel]

```python
reviews_per_year = df.groupby('review_year').size()
reviews_per_year.plot(kind='line', title='Reviews per Year', xlabel='Year', ylabel='Amount(Reviews)', color='skyblue', marker='o')
plt.xticks(reviews_per_year.index, rotation=0)
plt.show()
```

--------------------
Reviews by country

```python
top_countries = reviews_per_country.sort_values(ascending=False).head(6)

labels = [f"{country} ({count})" for country, count in zip(top_countries.index, top_countries.values)]

plt.figure(figsize=(6,6))
plt.pie(top_countries, labels=labels, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 11})
plt.title('Proportion of Reviews by Top 6 Countries')
plt.show()
```

###11. Conclusion [Daniel, Dylan, Antonio]

The analysis reveals that Ecuador leads with 346 reviews (48.2%), representing nearly half of all feedback. This strong local presence suggests that the hotel is highly recognized and frequently reviewed by domestic guests. The United States follows with 143 reviews (19.9%), indicating solid international interest, particularly from North American travelers. Germany (83 reviews, 11.6%), France (63 reviews, 8.8%), Spain (46 reviews, 6.4%), and the Netherlands (37 reviews, 5.2%) also contribute notable portions, reflecting steady engagement from European visitors.

Overall, these results highlight the hotel’s strong domestic reputation combined with a growing international appeal, especially among North American and European tourists. This balance suggests that the establishment successfully attracts both local and foreign markets, reinforcing its visibility and competitiveness in the hospitality sector.

The reviews per month show that June, November and December are the months with the most reviews. This indicates that the hotel is more popular during these months.

The reviews per year show that the number of reviews increased from 2022 to 2023, and then decreased in 2024 and 2025. This indicates that the hotel was gaining popularity in 2023 but has seen a decline in recent years.

Overall, the hotel attracts a diverse international and local clientele, with peak activity in the latter half of the year. The recent decline in reviews suggests a need for further investigation into potential factors impacting guest feedback and overall popularity.
